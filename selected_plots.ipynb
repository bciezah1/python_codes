# bar plot


```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a toy DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 25, 15, 30]}

df = pd.DataFrame(data)

# Plotting a bar plot
plt.bar(df['Category'], df['Values'])

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot from Toy Data')

# Display the plot
plt.show()
```

# distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating random data with a normal distribution
np.random.seed(42)  # Set seed for reproducibility
mean = 20
std_dev = 5
num_samples = 1000

random_data = np.random.normal(mean, std_dev, num_samples)

# Creating a DataFrame
df = pd.DataFrame({'Values': random_data})

# Plotting a histogram
plt.hist(df['Values'], bins=30, color='skyblue', edgecolor='black', density=True)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Histogram of Random Data with Normal Distribution')

# Display the plot
plt.show()

```

# correlation plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Generating random data for demonstration
np.random.seed(42)
data = {'A': np.random.normal(0, 1, 100),
        'B': np.random.normal(0, 1, 100),
        'C': np.random.normal(0, 1, 100)}

df = pd.DataFrame(data)

# Calculating the correlation matrix
correlation_matrix = df.corr()

# Plotting the correlation matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Adding correlation coefficients to the plot
for i in range(len(correlation_matrix)):
    for j in range(len(correlation_matrix)):
        if i != j:
            corr_coefficient, _ = pearsonr(df.iloc[:, i], df.iloc[:, j])
            plt.text(j + 0.5, i + 0.5, f'{corr_coefficient:.2f}', ha='center', va='center', color='black')

# Adding labels and title
plt.title('Correlation Plot with Coefficients')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample DataFrame with 100 points, groups, and size values
np.random.seed(42)
data = {'X': np.random.rand(100) * 10,
        'Y': np.random.rand(100) * 20,
        'Group': np.random.choice(['A', 'B', 'C'], size=100),
        'Size': np.random.randint(5, 50, size=100)}
df = pd.DataFrame(data)

# Mapping colors to groups
colors = {'A': 'red', 'B': 'green', 'C': 'blue'}
df['Color'] = df['Group'].map(colors)

# Plotting a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(df['X'], df['Y'], c=df['Color'], s=df['Size'], alpha=0.7, edgecolors='black')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Groups and Size')

# Display the legend
for group, color in colors.items():
    plt.scatter([], [], c=color, label=group)

plt.legend()

# Display the plot
plt.show()
```
